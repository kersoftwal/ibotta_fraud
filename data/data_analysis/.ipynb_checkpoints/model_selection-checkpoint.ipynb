# Model Selection


For the moment we're going to try clustering algorithms only on one-length sequences of transactions (single transactions) and gather outlying (suspicious) transactions. Then we will hopefully print a graph with the umap library. Nonetheless, we're going to consider one tenth of the whole dataset to get just a working pipeline on a low-resources machine.

Since we have no labels to understand whether the outliers are effectively retrieved, we're going to run several clustering algorithms and observe all red outlying stains appearing on the graphs. According to the most common clusters, how the algorithms work and the time spent for the execution, we're going to select the most appropriate model.

The most interesting three algorithms that will be taken into consideration are:
* hdbscan, the enhanced faster version of dbscan, chosen because performs really well in the general context of unsupervised learning and locality about regions detection, much used also for anomaly detection.
* isolation forests, very fast and effective, it is typically adopted for the specific task of anti-money laundering.
* autoencoders, famous for being really effective at understanding the structure of the data, used also in anomaly detection by comparing the original data and the rebuilt. 

The other algorithms will be considered for clustering region validation, namely k-nearest neighbors, local outlier factors, principal component analysis. 

```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

import eif 
from hdbscan import HDBSCAN
from pyod.models.auto_encoder import AutoEncoder
from pyod.models.iforest import IForest
from pyod.models.knn import KNN
from pyod.models.lof import LOF
from pyod.models.pca import PCA
```

```python
with open('crafted/df_cleaned.pickle', 'rb') as handle:
    df = pickle.load(handle)
with open('crafted/sequences_1trans.pickle', 'rb') as handle:
    scaled_df = pickle.load(handle)
with open('crafted/umap_1trans.pickle', 'rb') as handle:
    umap_1trans_df = pickle.load(handle)
```

```python
print(type(scaled_df))
scaled_df.shape
```

```python
data = scaled_df[:100000,:]
```

## K Nearest Neighbors

```python
knn = KNN(
    contamination = 0.03,
    #n_neighbors = 15,
    method = 'median'
)
%time knn.fit(data)
```

```python
knn_outliers = np.where(knn.labels_ == 1)[0]
knn_inliers = np.where(knn.labels_ == 0)[0]

print(len(knn_inliers))
print(len(knn_outliers))

plt.figure(figsize=(12,8))
plt.scatter(*umap_1trans_df.T, s=15, linewidth=0, c='gray', alpha=0.25)
plt.scatter(*umap_1trans_df[knn_outliers].T, s=15, linewidth=0, c='red', alpha=0.5)
plt.show()
```

```python
df.iloc[knn_outliers,:].describe()
```

```python
df.iloc[knn_inliers,:].describe()
```

## Local Outlier Factor algo

```python
lof = LOF(
    contamination = 0.03
)
%time lof.fit(data)
```

```python
lof_outliers = np.where(lof.labels_ == 1)[0]
lof_inliers = np.where(lof.labels_ == 0)[0]

print(len(lof_inliers))
print(len(lof_outliers))

plt.figure(figsize=(12,8))
plt.scatter(*umap_1trans_df.T, s=15, linewidth=0, c='gray', alpha=0.25)
plt.scatter(*umap_1trans_df[lof_outliers].T, s=15, linewidth=0, c='red', alpha=0.5)
plt.show()
```

```python
df.iloc[lof_outliers,:].describe()
```

```python
df.iloc[lof_inliers,:].describe()
```

## PCA

```python
pca = PCA(
    contamination = 0.1,
    whiten = True,
    random_state = 0
)
%time pca.fit(data)
```

```python
pca_outliers = np.where(pca.labels_ == 1)[0]
pca_inliers = np.where(pca.labels_ == 0)[0]

print(len(pca_inliers))
print(len(pca_outliers))

plt.figure(figsize=(12,8))
plt.scatter(*umap_1trans_df.T, s=15, linewidth=0, c='gray', alpha=0.25)
plt.scatter(*umap_1trans_df[pca_outliers].T, s=15, linewidth=0, c='red', alpha=0.5)
plt.show()
```

```python
df.iloc[pca_outliers,:].describe()
```

```python
df.iloc[pca_inliers,:].describe()
```

## HDBSCAN

```python
%time clusterer = HDBSCAN(min_cluster_size=15).fit(data)
```

```python
sns.distplot(clusterer.outlier_scores_[np.isfinite(clusterer.outlier_scores_)], rug=False)
```

### HDBSCAN using GLOSH algo

```python
threshold = pd.Series(clusterer.outlier_scores_).quantile(0.966)
hdbscan_glosh_outliers = np.where(clusterer.outlier_scores_ > threshold)[0]
hdbscan_glosh_inliers = np.where(clusterer.outlier_scores_ <= threshold)[0]

print(len(hdbscan_glosh_inliers))
print(len(hdbscan_glosh_outliers))

plt.figure(figsize=(12,8))
plt.scatter(*umap_1trans_df.T, s=15, linewidth=0, c='gray', alpha=0.25)
plt.scatter(*umap_1trans_df[hdbscan_glosh_outliers].T, s=15, linewidth=0, c='red', alpha=0.5)
plt.show()
```

### HDBSCAN using labels

```python
hdbscan_outliers = np.where(clusterer.labels_ == -1)[0]
hdbscan_inliers = np.where(clusterer.labels_ > -1)[0]

print(len(hdbscan_inliers))
print(len(hdbscan_outliers))

plt.figure(figsize=(12,8))
plt.scatter(*umap_1trans_df.T, s=15, linewidth=0, c='gray', alpha=0.25)
plt.scatter(*umap_1trans_df[hdbscan_outliers].T, s=15, linewidth=0, c='red', alpha=0.5)
plt.show()
```

```python
df.iloc[hdbscan_outliers,:].describe()
```

```python
df.iloc[hdbscan_inliers,:].describe()
```

### Isolation Forests


```python
isolation_forests = IForest(
    contamination=0.03,
    random_state=0
)

%time isolation_forests.fit(data)
```

```python
isolation_forests_outliers = np.where(isolation_forests.labels_ == 1)[0]
isolation_forests_inliers = np.where(isolation_forests.labels_ == 0)[0]

print(len(isolation_forests_inliers))
print(len(isolation_forests_outliers))

plt.figure(figsize=(12,8))
plt.scatter(*umap_1trans_df.T, s=15, linewidth=0, c='gray', alpha=0.25)
plt.scatter(*umap_1trans_df[isolation_forests_outliers].T, s=15, linewidth=0, c='red', alpha=0.5)
plt.show()
```

```python
df.iloc[isolation_forests_outliers,:].describe()
```

```python
df.iloc[isolation_forests_inliers,:].describe()
```

### Extended isolation forests

```python
extended_if = eif.iForest(data, ntrees=200, sample_size=256, ExtensionLevel=1)
%time eif_outliers_probs = extended_if.compute_paths(X_in=data)
```

```python
sns.distplot(eif_outliers_probs, kde=True, color="b")
plt.show()
```

```python
n_outliers = 3000
eif_outliers = eif_outliers_probs.argsort()[-n_outliers:][::-1]
eif_inliers = eif_outliers_probs.argsort()[:-n_outliers][::-1]

print(len(eif_inliers))
print(len(eif_outliers))

plt.figure(figsize=(12,8))
plt.scatter(*umap_1trans_df.T, s=15, linewidth=0, c='gray', alpha=0.25)
plt.scatter(*umap_1trans_df[eif_outliers].T, s=15, linewidth=0, c='red', alpha=0.5)
plt.show()
```

## Autoencoders

```python
ae = AutoEncoder(
    hidden_neurons = [17,12,12,17],
    epochs = 15,
    contamination = 0.1,
    random_state = 0
)
%time ae.fit(data)
```

```python
ae_outliers = np.where(ae.labels_ == 1)[0]
ae_inliers = np.where(ae.labels_ == 0)[0]

print(len(ae_inliers))
print(len(ae_outliers))

plt.figure(figsize=(12,8))
plt.scatter(*umap_1trans_df.T, s=15, linewidth=0, c='gray', alpha=0.25)
plt.scatter(*umap_1trans_df[ae_outliers].T, s=15, linewidth=0, c='red', alpha=0.5)
plt.show()
```

```python
df.iloc[ae_outliers,:].describe()
```

```python
df.iloc[ae_inliers,:].describe()
```
